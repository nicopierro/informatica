```python
from neo4j import GraphDatabase, basic_auth
import pandas as pd
pd.options.display.max_rows = 999
```

```python
driver = GraphDatabase.driver("bolt://52.87.238.189:7687", auth=basic_auth("neo4j", "pleasure-thunder-casualties"))
```

Prendere in considerazione il database Movies e scrivere in un file word le istruzioni Cypher per estrarre le seguenti informazioni:

```python
# 0. tutto il database
cypher_query = '''
MATCH (a)-[r]-(b) 
RETURN a, r, b
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df.head()
```

```python
# 1. i film
cypher_query = '''
MATCH (m:Movie) 
RETURN m
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df.head()
```

```python
# 2. i titoli dei film
cypher_query = '''
MATCH (m:Movie) 
RETURN m.title
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df.head()
```

```python
# 3. i titoli dei film in ordine alfabetico
cypher_query = '''
MATCH (m:Movie) 
RETURN m.title 
ORDER BY m.title
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 4. i titoli dei film dalla Z alla A
cypher_query = '''
MATCH (m:Movie) 
RETURN m.title 
ORDER BY m.title DESC
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 5. il numero di film
cypher_query = '''
MATCH (m:Movie) 
RETURN COUNT(m)
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 6. i film presentati tra il 1995 e il 2000 (estremi inclusi)
cypher_query = '''
MATCH (m:Movie) 
WHERE m.released >= 1995 AND m.released <= 2000 
RETURN m
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 7. il numero di film presentanti in ogni anno. Chiamare il numero di film "totale". Ordinare sul numero di film in ordine decrescente
cypher_query = '''
MATCH (m:Movie) 
RETURN m.released, COUNT(m) AS totale 
ORDER BY totale DESC
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 8. tutte le persone coinvolte nel film "Top Gun"
cypher_query = '''
MATCH (p:Person)-[r]-(m:Movie {title: "Top Gun"}) 
RETURN p.name, type(r)
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 9. gli attori del film "Top Gun"
cypher_query = '''
MATCH (m:Movie {title: 'Top Gun'})-[a:ACTED_IN]-(p:Person) 
RETURN p.name
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 10. i nomi degli attori del film "Top Gun" ordinati in ordine alfabetico
cypher_query = '''
MATCH (m:Movie {title: 'Top Gun'})<-[a:ACTED_IN]-(p:Person) 
RETURN p.name 
ORDER BY p.name
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 11. i nomi degli attori del film "Top Gun" ordinati in ordine alfabetico e il ruolo che hanno recitato nel film
cypher_query = '''
MATCH (m:Movie {title: 'Top Gun'})<-[a:ACTED_IN]-(p:Person) 
RETURN p.name, a.roles
ORDER BY p.name
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 12. il numero di attori del film "Top Gun"
cypher_query = '''
MATCH (m:Movie {title: 'Top Gun'})<-[a:ACTED_IN]-(p:Person) 
RETURN COUNT(p)
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 13. il regista e l'autore di "Top Gun"
cypher_query = '''
MATCH (m:Movie {title: 'Top Gun'})<-[a:WROTE|DIRECTED]-(p:Person) 
RETURN p.name, type(a)
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 14. l'elenco dei titoli dei film con i registi e gli attori che vi hanno preso parte
cypher_query = '''
MATCH (m:Movie)<-[a:DIRECTED|ACTED_IN]-(p:Person) 
RETURN m.title, p.name, type(a)
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```

```python
# 15. i film in cui ha recitato Tom Cruise e i rispettivi registi
cypher_query = '''
MATCH (tom:Person {name: "Tom Cruise"})-[:ACTED_IN]->(m:Movie)<-[:DIRECTED]-(p:Person) 
RETURN m.title, p.name
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
df = pd.DataFrame(result)
df
```
